Sets supports the following operations:
○ Intersection
○ Union
○ Difference
○ Symmetric Difference


```python
# Create a set of the following sentences
# 'ravi and akash are both good in statistics but akash is better in Ml modelling'
# 'ravi and akshay are equally good at python programing but ravi is better in statistics'

# And find:
# The common words in sent1 and sent2
# The set words present in both sent1 and sent2
# Words in sent1 which are not present in sent2
# Words in sent2 which are not present in sent1
# All the words in both sentences except the intersection words
```

```python
set1 = set('ravi and akash are both good in statistics but akash is better in Ml modelling'.split())
set1 
```

```python
set2 = set('ravi and akshay are equally good at python programing but ravi is better in statistics'.split())
set2
```

```python
set1.intersection(set2)
```

```python
set1.union(set2)
```

```python
set1 - set2
```

```python
set1.difference(set2)
```

```python
set2 - set1
```

```python
set1.symmetric_difference(set2)
```

```python
# Write a program to take age as input. If the driver is above eighteen check for the license. if 
# driver is above eighteen and has licence he is eligible to drive, else he is not. 

age = int(input('Enter your age: '))

if age >= 18:
    lic = input('Do you have license Y/N: ')
    if lic.upper()=='Y':
        print('You can drive')
    elif lic.upper()=='N':
        print('Not eligible to drive')
    else:
        print('Enter appropriate input')
else:
    print('Not old enough to drive')
```

```python
# Q. Take the marks and fees as the input. if marks are greater than 80% he/she is eligible for 
# scholarship. For boys 30% scholarship and for girls 50% scholarship. Print the total fees paid 
# by the student.

marks = int(input('Enter marks: '))
fees = int(input('Enter fees: '))

if marks >= 80:
    gender = input('Enter Male or Female M/F: ')
    if gender.lower() == 'm':
        fees = fees - fees*0.3
        print('Your fees: ',fees)
    else:
        fees = fees - fees*0.5
        print('Your fees: ',fees)
else:
    print('Your fees: ',fees)
```
